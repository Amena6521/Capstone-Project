## Identify and Analyze Duplicates


### Task 1: Identify Duplicate Rows
1. Count the number of duplicate rows in the dataset.
3. Display the first few duplicate rows to understand their structure.


```python
## Write your code here
# 1. Identify duplicate rows (considering all columns)
duplicates_mask = df.duplicated(keep=False)  # keep=False marks all duplicates as True

# 2. Count the number of duplicate rows
num_duplicates = duplicates_mask.sum()
print(f"Number of duplicate rows: {num_duplicates}")

# 3. Display the first few duplicate rows to understand their structure
duplicate_rows = df[duplicates_mask]
print("First few duplicate rows:")
print(duplicate_rows.head())
```

### Task 2: Analyze Characteristics of Duplicates
1. Identify duplicate rows based on selected columns such as MainBranch, Employment, and RemoteWork. Analyse which columns frequently contain identical values within these duplicate rows.
2. Analyse the characteristics of rows that are duplicates based on a subset of columns, such as MainBranch, Employment, and RemoteWork. Determine which columns frequently have identical values across these rows.
   


```python
## Write your code here
import pandas as pd



# Step 1: Identify duplicates based on subset of columns
duplicates_mask = df.duplicated(subset=['MainBranch', 'Employment', 'RemoteWork'], keep=False)

# Step 2: Extract duplicate rows
duplicate_rows = df[duplicates_mask]

# Step 3: Group by the subset columns
grouped = duplicate_rows.groupby(['MainBranch', 'Employment', 'RemoteWork'])

# Step 4: For each group, check which columns have a single unique value
def consistent_columns(group):
    return group.nunique() == 1

# Apply the function to each group and collect results
consistency_df = grouped.apply(consistent_columns)

# Step 5: Summarize across groups which columns are always consistent
# For each column, check if it is consistent in all groups
consistent_across_all_groups = consistency_df.groupby(level=0).all()

print(consistent_across_all_groups)
```

### Task 3: Visualize Duplicates Distribution
1. Create visualizations to show the distribution of duplicates across different categories.
2. Use bar charts or pie charts to represent the distribution of duplicates by Country and Employment.


```python
## Write your code here

# Step 1: Identify duplicate rows
duplicates_mask = df.duplicated(keep=False)
duplicate_rows = df[duplicates_mask]

# Step 2: Count duplicates by Country
country_counts = duplicate_rows['Country'].value_counts()

# Step 3: Count duplicates by Employment
employment_counts = duplicate_rows['Employment'].value_counts()

# Plotting duplicates distribution by Country
plt.figure(figsize=(10, 5))
country_counts.plot(kind='bar')
plt.title('Distribution of Duplicates by Country')
plt.xlabel('Country')
plt.ylabel('Number of Duplicate Rows')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plotting duplicates distribution by Employment
plt.figure(figsize=(8, 4))
employment_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Duplicates by Employment')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

### Task 4: Strategic Removal of Duplicates
1. Decide which columns are critical for defining uniqueness in the dataset.
2. Remove duplicates based on a subset of columns if complete row duplication is not a good criterion.


```python
## Write your code here


# Step 1: Explore uniqueness of columns
print(df.nunique())  # Number of unique values per column

# For example, check if combination of columns uniquely identifies rows
print(df.duplicated(subset=['MainBranch', 'Employment', 'RemoteWork']).sum())

# Step 2: Remove duplicates based on subset of columns
df_no_duplicates = df.drop_duplicates(subset=['MainBranch', 'Employment', 'RemoteWork'], keep='first')

print(f"Original dataset size: {len(df)}")
print(f"Dataset size after removing duplicates: {len(df_no_duplicates)}")
```
